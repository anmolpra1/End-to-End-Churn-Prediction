```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
!pip install imbalanced-learn

from imblearn.combine import SMOTEENN # for upsampling and ENN
```

**Reading CSV**

```python
df = pd.read_csv("C:\\customer_churn_analysis\\anaconda_projects\\ea39725e-e0be-4238-a44f-a521c44e9093\\tel_churn.csv")

df.head()
```

```python
df.drop('Unnamed: 0', axis = 1)
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y                                     
```

**Train Test Split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

## Decision Tree Classifier


```python
model_dt = DecisionTreeClassifier(
    criterion="gini",
    random_state=100,
    max_depth=6,
    min_samples_leaf=8
)

# Controlling overfitting

# Making model more generalizable

# Adding bias to reduce variance
```

```python
model_dt.fit(x_train, y_train)
```

```python
y_pred = model_dt.predict(x_test)

y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

As you can see that the accuracy is quite low, and as it's an imbalanced dataset, we shouldn't consider Accuracy as our metrics to measure the model, as Accuracy is cursed in imbalanced datasets.
Hence, we need to check recall, precision & f1 score for the **minority class**, and it's quite evident that the precision, recall & f1 score is too low for Class 1, i.e. churned customers.
Hence, moving ahead to call **SMOTEENN** (UpSampling + ENN)

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x, y)  # Changed fit_sample to fit_resample
```
